# Bar chart

```python
from matplotlib import pyplot as plt
import numpy as np
year=[2008,2009,2010,2011,2012,2013,2014,2015,2016,2017,2018]
toyota_sales=[18884964,15555654,2155464,215461,4851122,1545595,1595666,1585488,5545564,2129212,5299334]
x_pos=np.arange(len(year))

plt.bar(x_pos,toyota_sales,color='#623aa2',edgecolor='k',label='Toyota',alpha=0.4)
plt.xticks(x_pos,year,rotation=30)
plt.title('Toyota car sales')
plt.xlabel('year')
plt.ylabel('no of units sold')
plt.show()
```

## Clustered Bar chart



```python
year=[2010,2011,2012,2013,2014,2015]
honda_sales=[5,6,2,8,9,15]
toyota_sales=[10,2,3,5,6,12]
hyundai_sales=[3,2,4,6,11,10]
x_pos=np.arange(len(year))
width=0.25
plt.bar(x_pos-width,honda_sales,color='r',edgecolor='k',label='Honda',width=width)
plt.bar(x_pos,toyota_sales,color='b',edgecolor='k',label='Toyota',width=width)
plt.bar(x_pos+width,hyundai_sales,color='c',edgecolor='k',label='Hyundai',width=width)
plt.xticks(x_pos,year,rotation=30)
plt.title('Car sales')
plt.xlabel('Year')
plt.ylabel('No of unit sold')
plt.legend()
plt.show()
```

```python
import csv
with open(r'H:\python course\Python Matplotlib - Udemy\AirPollution.csv') as input_file:
    csv_file=csv.reader(input_file,delimiter=',')  # object iterate over the csv file
    Header=next(csv_file)
    country=[]
    index=[]
    for row in csv_file:
        country.append(row[0])
        index.append(int(row[1]))

```
