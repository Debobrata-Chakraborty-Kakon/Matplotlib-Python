```python
%matplotlib notebook
from matplotlib import pyplot as plt
from matplotlib import animation
import numpy as np
```

```python
fig1 = plt.figure()
ax1 = plt.subplot(111)
line1,  = ax1.plot([],[])
ax1.set_xlim(0,2)
ax1.set_ylim(-1.5,1.5)
x = np.linspace(0,2,1000)
 
def init():
    ax1.set_xlabel('x')
    ax1.set_ylabel('sin(x)')
    ax1.set_title('sine curve')
    line1.set_data([],[])
    return line1,
 
def animate(i):
    y = np.sin(2 * np.pi * (x -  0.01*i))
    line1.set_data(x,y)
    return line1,
 
anim1 = animation.FuncAnimation(fig1,animate,init_func=init,frames=100,interval=10,repeat=False)
 
plt.show()
```

```python
from matplotlib import pyplot as plt
from matplotlib import animation
import numpy as np

fig3=plt.figure()
ax3=plt.axes()
ax3.set_xlim(0,20)
ax3.set_ylim(0,10)

def animate(i):
    x=[]
    y=[]
    
    f= open(r'H:\python course\Python Matplotlib - Udemy\Sample.txt','r')
    lines=f.read().split('\n')
    
    for line in lines:
        if len(line)>1:
            x1,y1=line.split(',')
            x.append(float(x1))
            y.append(float(y1))
    ax3.clear()
    ax3.plot(x,y,marker='o')

anim3=animation.FuncAnimation(fig3,animate,interval=1000)
plt.show()
```
