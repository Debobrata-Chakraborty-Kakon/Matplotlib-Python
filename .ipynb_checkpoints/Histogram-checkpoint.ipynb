## Frequency Distribution and Histogram


```python
from matplotlib import pyplot as plt
import numpy as np

input_data=np.random.randint(2,13,20)
input_data
```

```python
plt.hist(input_data,bins=20,color='r',alpha=0.5)
plt.title('Sum of 2 dice rolls')
plt.xlabel('bins')
plt.ylabel('no of outcomes')
plt.show()
```

```python
input_data2=np.random.rand(10**4)
plt.hist(input_data2,bins=100,color='r',log=True,alpha=0.5)
plt.title('Frequency distribution of random numbers')
plt.show()
```
