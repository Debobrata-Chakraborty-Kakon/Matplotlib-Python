```python
from matplotlib import pyplot as plt
plt.style.use('classic')
x=[22,18,13,10]
labels=['maths','physics','chemistry','english']
plt.pie(x,labels=labels,colors=['b','r','g','c'],autopct='%1.1f%%') #autopct -% value
plt.show()
```

```python
x=[22,18,13,10]
labels=['maths','physics','chemistry','english']
colors=['r','b','m','c']
plt.pie(x,labels=labels,colors=colors,explode=[0,0,.2,0],autopct='%1.1f%%',startangle=45,counterclock=True,shadow=True,
           wedgeprops={"edgecolor":"k",'linewidth':1,'linestyle':'-'}) #autopct -% value
plt.title('Result')
plt.legend()
plt.show()
```
